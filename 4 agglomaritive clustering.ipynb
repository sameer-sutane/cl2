```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data using make_blobs
X, y = make_blobs(n_samples=150, centers=3, cluster_std=0.6, random_state=42)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o', s=30, label='Data Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Sample Data for Clustering')
plt.legend()
plt.show()
```

```python
# Create an instance of AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Fit the model and predict the cluster labels
y_pred = agg_clustering.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_pred, marker='o', s=30, cmap='viridis', label='Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform hierarchical/agglomerative clustering and generate the linkage matrix
linkage_matrix = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
